```python
import random
import math
import sys
import re
import os
import operator

import simpy
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
from matplotlib.colors import ListedColormap
import numpy as np
from scipy.stats import norm

from Node import Node, EnergyProfile
from Gateway import Gateway, DownlinkPacket
from TransmissionInterface import AirInterface
from LoRaParameters import LoRaParameters
from External import *
from Server import Server
from config import *


sim_time =  30 * MINUTE_TO_MS
times = [5000]
thresholds = [45, 60, 75]

# thresholds = [10, 20, 30]
num_nodes = len(range(-CORD,CORD,2))**2
transmission_status = np.zeros(num_nodes)


for time in times:
    for threshold in thresholds: 
        nodes = []
        gateways = []
        sim_env = simpy.Environment()
        external = TempExternal(sim_env)
        air = AirInterface(sim_env)
        gateways.append(Gateway(0, 0, 0,  sim_env))
        server = Server()
        
        n_id = 0
    

        for i in range(-CORD,CORD,5):
            for j in range(-CORD,CORD,5):
                

                x = j * GRID
                y = i * GRID
                node = Node(n_id, EnergyProfile(3.3), LoRaParameters(i%Gateway.NO_CHANNELS, sf = 12, tp=21), x,
                      y,gateways, 20, server, air, sim_env , external = external, status = transmission_status)
                nodes.append(node)
                n_id += 1
                sim_env.process(node.run(time = time, threshold = threshold))
        sim_env.process(external.update())
        sim_env.run(sim_time)
        
        T_rate = np.zeros(( len(nodes),int(sim_time/time)))
        S_rate= np.zeros(( len(nodes),int(sim_time/time)))
        for i in range(len(nodes)):
            T = np.array(nodes[i].true_history)
            dT = np.absolute(np.array(nodes[i].value_history) - T)
            T_rate[i, :] = 1 - np.divide(dT, T)
            S_rate[i, :] = np.array(nodes[i].success_history)
        np.save('Trate_%d_%d_1.npy' %(time, threshold), T_rate)
        np.save('Srate_%d_%d_1.npy' %(time, threshold), S_rate)

print(n_id)
```

```python
cor = np.zeros( (2, len(nodes)))
for i in range(len(nodes)):
    cor [0,i] = nodes[i].x
    cor [1,i] = nodes[i].y
np.save('coordinate.npy', cor)    # .npy extension is added if not given
```
